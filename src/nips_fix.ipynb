```python
def increase_uid(row,i):
    species = row['uid']
    spec = species.split('_')
    uid_num = int(spec[0]) + (i*86)
    spec[0] = str(uid_num)
    spec_to = '_'.join(spec)
    return spec_to

# window_2['uid'] = window_2.apply(lambda row: increase_uid(row),axis = 1)
# window_2
```

```python
def increase_time(row,i):
    time = row['temporal_frame_start_times']
    time += (i*2)
    return time

# window_2['temporal_frame_start_times'] = window_2.apply(lambda row: increase_time(row),axis = 1)
# window_2
```

```python
file_1.shape[0]//86
```

```python
nips_fix = []

for i in range(6):
    print(f'start {i * 86} end {(i+1)*86}')
    curr_win = file_1.iloc[i * 86:(i+1)*86,:].copy(True)
    curr_win['uid'] = curr_win.apply(lambda row: increase_uid(row,i),axis = 1)
    curr_win['temporal_frame_start_times'] = curr_win.apply(lambda row: increase_time(row,i),axis = 1)
    # print(type(curr_win))
    nips_fix.append(curr_win)

nips_fix = pd.concat(nips_fix)
```
