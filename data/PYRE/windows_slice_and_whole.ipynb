```python
# import Load_data_functions
import pandas as pd
import os
import pickle
import librosa
from librosa import display
import matplotlib.pyplot as plt
```

```python
# load_pckl = Load_data_functions.load_pyrenote_dataset(,'Mixed_Bird-20220126T212121Z-003',44100,86,2048,1024)
# load_pckl

mel_dump_file = os.path.join('downsampled_Mixed_Bird-20220126T212121Z-003_bin_mel_dataset.pkl')
# if os.path.exists(mel_dump_file):

with open(mel_dump_file, "rb") as f:
    dataset = pickle.load(f)
file = dataset.get('uids')[13] 
print(file)
bird_0 = dataset.get('X')[13]
fig, ax = plt.subplots(figsize=(20,5))
spec3 = librosa.display.specshow(bird_0, hop_length = 2048,sr = 44100, y_axis='mel', x_axis='time') # 44100,86,2048,1024
# spec3.axtitle(title = dataset.get('uids')[0])
ax.set(title=file)
plt.savefig('Attila-cinnamomeus-206109.png')
plt.show()
```
